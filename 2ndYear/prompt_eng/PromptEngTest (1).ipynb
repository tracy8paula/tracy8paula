```python
# Requires: openai >= 1.60.0
# Run: pip install --upgrade openai

from openai import OpenAI
import time

# Create an OpenAI client (replace with your real key or env var)
client = OpenAI(api_key="Your OPENAI Key Here")

# Conversation state
# Start with a "system" instruction as a *string* (not a list)
system_prompt = "You are an intelligent assistant. Reply with 15-50 words." # this specifies the range!

# Keep every turn as a simple string in `history`
history = [f"System: {system_prompt}"]

# Interactive loop
print("Type your messages. Type 'bye' to exit. Type 'save' to save history. Type 'summary' for a recap.\n")

while True:
    user_msg = input("User: ").strip()
    if not user_msg:
        continue # this checks if input is empty and skips the rest of the loop until input is filled
    if user_msg.lower() == "bye":
        print("GPT-5: Goodbye ðŸ‘‹")
        break

    # saving conversation as a text file
    if user_msg.lower() == "save":
        with open("conversation_history.txt", "w", encoding="utf-8") as f:
            f.write("\n".join(history))
            print("Conversation saved to conversation_history.text\n")
            continue

    # summarising the conversation
    if user_msg.lower() == "summary":
        summary_prompt = "\n".join(history) + "\nAssistant: Please provide a concise summary of this conversation so far."
        try:
            response = client.responses.create(model="gpt-5",input = summary_prompt)
            summary = response.output_text.strip()
            print(f"GPT-5 (Summary): {summary}\n")
            history.append(f"Assistant (Summary): {summary}")
        except Exception as e:
            print("Error while summarising:", e)
        continue

    # Save the user input, the code is appended after it's collected and checked.
    history.append(f"User: {user_msg}")

    # Join the transcript into one big text block
    prompt = "\n".join(history)

    # error handling in the API
    reply = None
    for attempt in range(3):
        try:
            response = client.responses.create(model="gpt-5", input=prompt, 
                                               reasoning={"effort": "medium"}, text={"verbosity": "medium"}) # this is the method to send the conversation prompt to GPT-5
            reply = response.output_text.strip()
            break
        except Exception as e:
            print(f"Error (attempt {attempt+1}):", e)
            if attempt < 2:
                print("Retrying...\n")
                time.sleep(2) # waiting before retry
            else:
                print("Failed after 3 attempts. Exiting.")
                quit()

    # showing the assistant reply
    if reply:
        print(f"GPT-5: {reply}\n")
        history.append(f"Assistant: {reply}")
```
